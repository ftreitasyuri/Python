```python
import os
import datetime
import csv
import requests
import json

credentials = 'email@intelipost.com.br', 'password'
zendesk = 'https://intelipost.zendesk.com'
language = 'pt-br'


date = datetime.date.today()
backup_path = os.path.join(str(date), language)
# se não existir a pasta então crie
if not os.path.exists(backup_path):
    # comando para criar a pasta
    os.makedirs(backup_path)

log = []

# variavel endpoint recebe variavel zendesk e concatena com a api passando tudo para minusculo
endpoint = zendesk + '/api/v2/help_center/pt-br/articles.json'.format(locale=language.lower())
# respose faz um requests e autentica o end point mais as credenciais
response = requests.get(endpoint, auth=credentials)

while endpoint:
    response = requests.get(endpoint, auth=credentials)
    if response.status_code != 200:
        print('Failed to retrieve articles with error {}'.format(response.status_code))
        exit()
    data = response.json()

    for article in data['articles']:
       if article['body'] is None:
           continue
       title = '<h1>' + article['title'] + '</h1>'
       filename = '{id}.html'.format(id=article['id'])
       with open(os.path.join(backup_path, filename), mode='w', encoding='utf-8') as f:
     
           f.write(title + '\n' + article['body'])
       print('{id} copied!'.format(id=article['id']))


       
       log.append((filename, article['title'], article['author_id']))
       
    endpoint = data['next_page']

with open(os.path.join(backup_path, '_log.csv'), mode='wt', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow( ('File', 'Title', 'Author ID') )
    for article in log:
        writer.writerow(article)


# comando para testar se estamos recebendo os artigos da pagina
# por padrão só receberemos 30 artigos
# for article in data['articles']:
#     print(article['id'])

```
