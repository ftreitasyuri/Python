```python
# Importar a base de dados
import pandas as pd
import openpyxl as openpyxl
import win32com.client as win32
```

```python
tabela_venda = pd.read_excel('DataBase/Vendas.xlsx')
# Comando para fazer com que o pandas não oculte nenhuma coluna do database
pd.set_option('display.max_columns', None)
```

```python
print(tabela_venda)
```

# faturamento por loja

```python
faturamento = tabela_venda[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
faturamento
```

# quantidade de produtos vendidos por loja

```python
# Atribuindo à variável quantidade as duas colunas ID Loja e Quantidade e 
# agrupando por Id Loja e somando a quantidade de produtos vendidos

quantidade = tabela_venda[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()

quantidade
```

# ticket médio por produto em cada loja

```python
# Ticket médio é o faturamento dividido pela quantidade
ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio
```

# Enviar um email com o relatório por GMAIL

```python
import smtplib
import email.message
```

```python
# Código para enviar email pelo GMAIL

def enviar_email(): 
    corpo_email = f"""
    <p> Prezaos,</p>
    
    <p>Segue meu primeiro email automático</p>
    
    Faturamento:
    {faturamento.to_html()}
    
    Quantidade:
    {quantidade.to_html()}
    
    Ticket Médio:
    {ticket_medio.to_html()}
    
    <p>Abs</p>
    <p>Yuri</p>
    
    """

    msg = email.message.Message()
    msg['Subject'] = "Meu primeiro email automático"
    msg['From'] = 'remetente'
    msg['To'] = 'destinatario'
    password = 'senhaRemetente' # senha do e-mail from
    msg.add_header('Content-Type', 'text/html')
    msg.set_payload(corpo_email )

    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()
    # Login Credentials for sending the mail
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    print('Email enviado')
```

```python
enviar_email()
```
