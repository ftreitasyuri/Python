## Tratamente de dados faltantes continuação

```python
# Para ver o tamanho da váriavel Condominio
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Na variavel selecao estamos pegando do dataframe dados na coluna tipo tudo que é igual apartamento
# e que tem o condominio nulo
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# Eliminando da variável selecao tudo que esta com condominio nule e que é do tipo apartamento
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
# Atribuindo valor a todos os IPTU's vazios
# PRIMEIRO JEITO DE FAZER
# dados.fillna(0, inplace = True)
# SEGUNDO JEITO DE FAZER
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```
